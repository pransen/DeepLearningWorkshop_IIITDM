```python
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt


# Load image as grayscale
img = np.zeros((224, 224))
img[:, :112] = 255  # Top-left quadrant

plt.imshow(img, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')
plt.show()
```

## Prewitt and Sobel Kernels

### ðŸŸ¦ Prewitt Operator Kernels

[-1 0 1]

[-1 0 1]

[-1 0 1]


```python
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewitty = cv2.filter2D(img, -1, kernely)
plt.imshow(img_prewitty, cmap='gray')
plt.title('Vertical Edge Detection (Prewitt Y)')
plt.axis('off')
plt.show()
```

### ðŸŸ© Sobel Operator Kernels

**Horizontal Kernel (Gx):**

[-1 0 1]

[-2 0 2]

[-1 0 1]

```python
import cv2

# Vertical Sobel kernel
# Apply Sobel Y filter to detect vertical edges
sobel_y = cv2.Sobel(img, cv2.CV_64F, dx=1, dy=0, ksize=3)

# Convert to absolute and uint8 for display
sobel_y_abs = cv2.convertScaleAbs(sobel_y)

plt.imshow(sobel_y_abs, cmap='gray')
plt.title('Vertical Edge Detection (Sobel Y)')
plt.axis('off')
plt.show()
```
