# Image Thresholding

## ðŸ“Œ Image Thresholding

## 1. Introduction

Image thresholding is used to segment an image by converting it into a binary image. This means each pixel will either be black (0) or white (255), based on a comparison with a predetermined **threshold value**.

The core idea is to differentiate between foreground objects and the background.

## 2. The Thresholding Formula

For a given pixel intensity $I(x, y)$ at coordinates $(x, y)$ and a chosen threshold $T$, the output binary pixel $B(x, y)$ is calculated as follows:

$$
B(x, y) = \begin{cases}
255 & \text{if } I(x, y) > T \\
0 & \text{if } I(x, y) \le T
\end{cases}
$$

* **$I(x, y)$**: Original pixel intensity.
* **$T$**: The threshold value.
* **$B(x, y)$**: Output binary pixel intensity.



```python
import cv2
import matplotlib.pyplot as plt

# Read image in grayscale
image = cv2.imread('watershed_coins_01.jpg', cv2.IMREAD_GRAYSCALE)

# Apply binary thresholding
# Parameters: src, threshold value, max value, thresholding type
threshold_value = 127
max_value = 255
_, thresh_image = cv2.threshold(image, threshold_value, 127, cv2.THRESH_BINARY)

# Show original and thresholded image
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title("Original Grayscale")
plt.imshow(image, cmap='gray')

plt.subplot(1,2,2)
plt.title(f"Thresholded (Binary) - {threshold_value}")
plt.imshow(thresh_image, cmap='gray')
plt.show()
```

# Understanding the Image Histogram

---

An image histogram is a graphical representation of the distribution of intensity levels in a digital image. It plots the number of pixels for each intensity value found in the image. Essentially, it shows how many pixels have a certain brightness level.

What is an Image Histogram?

For a grayscale image, where pixel intensities typically range from 0 (black) to 255 (white), a histogram will have 256 bins (one for each intensity level). Each bin represents a specific intensity value, and its height indicates the **count** (or frequency) of pixels in the image that have that particular intensity.

* **X-axis:** Represents the pixel intensity values (e.g., 0 to 255 for an 8-bit image).
* **Y-axis:** Represents the number of pixels having that intensity value.


```python
# Calculate histogram
# hist = cv2.calcHist([image], [0], None, [256], [0, 256])

import numpy as np

hist = np.zeros(256)

for x in range(image.shape[0]):
  for y in range(image.shape[1]):
    hist[image[x][y]] += 1


# Convert to 1D array
hist = hist.ravel()

# Plot using bar graph
plt.figure(figsize=(8, 4))
plt.title("Grayscale Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.bar(range(256), hist, width=1.0, color='black')
plt.xlim([0, 256])
plt.grid(True)
plt.show()
```

```python
# Apply binary thresholding
# Parameters: src, threshold value, max value, thresholding type
threshold_value = 75
max_value = 255
_, thresh_image = cv2.threshold(image, threshold_value, max_value, cv2.THRESH_BINARY)

# Show original and thresholded image
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title("Original Grayscale")
plt.imshow(image, cmap='gray')

plt.subplot(1,2,2)
plt.title("Thresholded (Binary)")
plt.imshow(thresh_image, cmap='gray')
plt.show()
```
